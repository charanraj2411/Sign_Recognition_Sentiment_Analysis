```python
model2 = load_model('Disaster_Model.h5')
```

```python

corpus1 = []
X1 = re.sub(r'https?:\/\/.*[\r\n]*', '', sentence)
X1 = re.sub(r'[^a-zA-Z]', ' ', X1)
X1= re.sub(r'#', '', X1)
corpus1.append(X1)
print(corpus1)
```

```python
voc_size = 50000
onehot_repr1 = [one_hot(words,voc_size) for words in list(corpus1)]
print(onehot_repr1)
```

```python
Length_of_tweet=20
embedded_docs1 = pad_sequences(np.array(onehot_repr1),padding='pre',maxlen=Length_of_tweet)
print(embedded_docs1)
```

```python
y_pred = model2.predict_classes(embedded_docs1)
```

```python
print(y_pred)
```

```python
for i in y_pred:
    for j in i:
        if j==1:
            print('The person is in happy space')
        else:
            print('The person is in a sad mood')
```
